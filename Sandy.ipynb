```python
#first we do some imports and check the version of Py-ART for consistency
import pyart
from matplotlib import pyplot as plt
import netCDF4
import numpy as np
from copy import deepcopy
import copy
import scipy
from matplotlib import rc
from netCDF4 import num2date
import h5py
%matplotlib inline
print pyart.__version__
```

```python
filename = '/data/radar/maine/KGYX20121031_000150_V06'
radar = pyart.io.read(filename)
#43.6667° N, 70.2667° W
```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map( 'reflectivity', vmin = -8, vmax = 64, 
                 resolution = 'h', max_lat = 45, 
                 min_lat = 43, min_lon = -72, max_lon = -69)
m = myd.basemap
m.drawparallels(np.linspace(43, 45, 5),labels=[1,0,0,0])
m.drawmeridians(np.linspace(-72, -69, 4),labels=[0,0,0,1])
m.drawrivers()
m.drawmapscale(-71, 43.25, -71, 43.25, 100, barstyle='fancy', fontcolor='k', fillcolor1='b', fillcolor2='k')
myd.plot_point( -70.2667,43.6667, label_text = 'Portland', label_offset = (0.0,0.0) )

```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map( 'reflectivity', vmin = -8, vmax = 64, 
                 resolution = 'h', max_lat = 44, 
                 min_lat = 43.5, min_lon = -71, max_lon = -70)
m = myd.basemap
m.drawparallels(np.linspace(43.5, 44, 6),labels=[1,0,0,0])
m.drawmeridians(np.linspace(-71, -70, 5),labels=[0,0,0,1])
m.drawrivers()
m.drawmapscale(-70.75, 43.75, -70.75, 43.75, 10, barstyle='fancy', fontcolor='k', fillcolor1='b', fillcolor2='k')
myd.plot_point( -70.2667,43.6667, label_text = 'Portland', label_offset = (0.0,0.0) )


```

```python
rain_z = radar.fields['reflectivity']['data'].copy()
z_lin = 10.0**(radar.fields['reflectivity']['data']/10.)
rain_z = (z_lin/300.0)**(1./1.4)  #Z=300 R1.4
radar.add_field_like('reflectivity', 'rain_z',  rain_z, replace_existing = True)
radar.fields['rain_z']['units'] = 'mm/h'
radar.fields['rain_z']['standard_name'] = 'rainfall_rate'
radar.fields['rain_z']['long_name'] = 'rainfall_rate_from_z'
radar.fields['rain_z']['valid_min'] = 0
radar.fields['rain_z']['valid_max'] = 500





```

```python
myd = pyart.graph.RadarMapDisplay(radar)
fig = plt.figure(figsize = [18,10])
myd.plot_ppi_map( 'rain_z', vmin = 0, vmax = 20, 
                 resolution = 'h', max_lat = 44, 
                 min_lat = 43.5, min_lon = -71, max_lon = -70)
m = myd.basemap
m.drawparallels(np.linspace(43.5, 44, 6),labels=[1,0,0,0])
m.drawmeridians(np.linspace(-71, -70, 5),labels=[0,0,0,1])
m.drawrivers()
m.drawmapscale(-70.75, 43.75, -70.75, 43.75, 10, barstyle='fancy', fontcolor='k', fillcolor1='b', fillcolor2='k')
myd.plot_point( -70.2667,43.6667, label_text = 'Portland', label_offset = (0.0,0.0) )


```

```python
mapping = pyart.map.grid_from_radars(
    (radar,),  
    grid_shape=(1, 151, 151),
    grid_limits=((300, 300), 
                 (-30.*1000., 30.*1000.), 
                 (-30.*1000., 30.*1000.)),
    grid_origin = (43.6667,-70.2667),
    fields=['reflectivity', 'rain_z'],
    refl_field='reflectivity',
    max_refl=100.,
    copy_field_data=False)
```

```python
max_lat = 44
min_lat = 43.5 
min_lon = -71 
max_lon = -70

plt.figure(figsize = [15,10])
display = pyart.graph.GridMapDisplay(mapping)
display.plot_basemap(lat_lines=np.arange(min_lat,max_lat,.1),
                     lon_lines=np.arange(min_lon, max_lon, .1),
                     resolution='h')
display.plot_grid('reflectivity', vmin=-8, vmax=64)
xcf,ycf = display.basemap(-70.2667,43.6667)
display.basemap.plot(xcf,ycf,'ro')
plt.text(xcf+2000.,ycf+2000., 'Portland')
display.basemap.drawmapscale(-70.5, 43.75, -70.5, 43.75, 10, barstyle='fancy', fontcolor='k', fillcolor1='r', fillcolor2='k')
display.plot_colorbar()
```
